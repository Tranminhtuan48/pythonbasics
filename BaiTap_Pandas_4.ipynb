Câu 5: Người ta thường sử dụng chỉ số z (hay z-score) để xác định giá trị ngoại lệ (outlier). Bạn
hãy loại bỏ các hàng có chứa giá trị ngoại lệ ra khỏi dataframe.

```python
for i in ["Height", "Speed", "Distance", "Fatalities","Engines", "Injuries"]:
  mean = dt[i].mean()
  dt[i] = dt[i].fillna(mean)
dt1 = dt[["Height", "Speed", "Distance", "Fatalities","Engines", "Injuries"]]
z = np.abs(st.zscore(dt1))
dt1 = dt1[(z < 3).all(axis=1)]
dt1
```
