* Sử dụng chỉ số z (hay z-score) để xác định giá trị ngoại lệ (outlier), sau đó tiến hành loại bỏ ngoại lệ

```python
for i in ["Mthly_HH_Income", "Mthly_HH_Expense", "No_of_Fly_Members", "Emi_or_Rent_Amt","Annual_HH_Income", "No_of_Earning_Members"]:
  mean = dt[i].mean()
  dt[i] = dt[i].fillna(mean)
dt1 = dt[["Mthly_HH_Income", "Mthly_HH_Expense", "No_of_Fly_Members", "Emi_or_Rent_Amt","Annual_HH_Income", "No_of_Earning_Members"]]
z = np.abs(st.zscore(dt1))
dt1 = dt1[(z < 3).all(axis=1)]
dt1
```
