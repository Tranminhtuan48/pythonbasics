QUY TRÌNH PHÂN TÍCH ANOVA MỘT CHIỀU

Kiểm tra các giả định:

- Các mẫu độc lập với nhau
- Các mẫu có cùng phương sai
- Các mẫu đều tuân theo phân phối chuẩn

Xây dựng giả thuyết

$$H_{0}: \mu_{0} = \mu_{1} = ... = \mu_{n}$$
$$H_{a}: \text{Có tối thiểu một cặp giá trị trung bình có giá trị khác nhau}$$

Giá trị thống kê F được sử dụng trong việc ra quyết định của phân tích ANOVA một chiều. Công thức tính như sau:

$$F = \frac{MS_{B}}{MS_{W}}$$

với:

- $MS_{B} = \frac{\sum_{k}n_{k}(\overline{\rm x_{k}} - \overline{\rm x})^{2}}{k - 1}$
- $MS_{W} = \frac{\sum_{k}n_{i,k}( x_{i,k} - \overline{\rm x})^{2}}{k - 1}$

Loại bỏ giả thuyết $H_{0}$ khi chỉ số F > chỉ số F-critical (hoặc giá trị p tương ứng < 0.05)

```python
#Diet1
df1=dataset[dataset["Diet"] == 1] 
df1
```

###**Phát biểu giả thuyết:** (với mức ý nghĩa α = 0,05)
$$H_{0} : \mu_{pre_w} = \mu_{W_6}$$

$$H_{A} : \text{Có tối thiểu một cặp giá trị trung bình có giá trị khác nhau} $$

Giả thuyết gốc: "Không có sự khác biệt về giá trị trung bình của cân nặng trước và sau 6 tuần ăn kiêng"

Giả thuyết thay thế: "Có sự khác biệt về giá trị trung bình của cân nặng trước và sau 6 tuần ăn kiêng"

**Tiến hành kiểm định các điều kiện *Phân phối chuẩn, phương sai*:**

Phát biểu giả thuyết với mức ý nghĩa *α = 0.05*:
- Null: "Nhóm pre_weight và weight6week được lấy từ phân phối chuẩn và không có sự khác biệt về phương sai giữa các nhóm" 
- Giả thuyết thay thế: "Nhóm pre_weight và weight6week không được lấy từ phân phối chuẩn và có sự khác biệt về phương sai giữa các nhóm"

```python
#phan phoi chuan pre.weight
p_value = stats.shapiro(df1["pre.weight"])[1]
print(f"P_value của pre_weight: {p_value}")
if p_value < 0.05:
  print(f"Có bằng chứng cho thấy pre_weight không tuân theo phân phối chuẩn")
else:
  print(f"Không có bằng chứng cho thấy nhóm pre_weight không tuân theo phân phối chuẩn")
#phan phoi chuan weight6weeks
p_value = stats.shapiro(df1["weight6weeks"])[1]
print(f"P_value của weight6weeks: {p_value}")
if p_value < 0.05:
  print(f"Có bằng chứng cho thấy weight6weeks không tuân theo phân phối chuẩn")
else:
  print(f"Không có bằng chứng cho thấy weight6weeks không tuân theo phân phối chuẩn")
#phuong sai
stat_value ,p_value = stats.levene(df1["pre.weight"], df1["weight6weeks"])
print(f"P value: {p_value}")
if p_value > 0.05:
    print(f"Có bằng chứng cho thấy sự đồng nhất phương sai giữa các nhóm")
else:
    print(f"Không có bằng chứng cho thấy có sự đồng nhất phương sai giữa các nhóm")
```

**Các nhóm đều độc lập, có cùng phương sai và tuân theo phân phối chuẩn nên ta tiến hành kiểm định ANOVA:**

```python
f_stat, p_value = stats.f_oneway(df1["pre.weight"], df1["weight6weeks"])
print(f"P value: {p_value}")
if p_value < 0.05:
    print(f"Không có bằng chứng để bác bỏ giả thuyết H0")
else:
    print(f"Có bằng chứng để bác bỏ giả thuyết H0")
```

Vậy có bằng chứng cho thấy có sự khác biệt về cân nặng trung bình sau 6 tuần ăn kiêng theo chế độ 1

```python
#tính giá trị trung bình độ lệch cân nặng để so sánh
st1 = df1["pre.weight"].mean() - df1["weight6weeks"].mean()
print(st1)
```
