```python
import scipy.stats as ss
import numpy as np
import statsmodels.stats as sm
```

```python
data = np.random.uniform(size = 30)
ks_statistic, p_value = ss.kstest(data, "uniform")
print(ks_statistic, p_value)
```

```python
# Kiểm định có phải là phân phối poisson
data = np.random.poisson(10, size = 30)
ks_statistic, p_value = ss.kstest(data, 'poisson', args=(10,0))
print(ks_statistic, p_value)
```

```python
# Kiểm định có phải là phân phối chuẩn
data = np.random.normal(size = 30)
ss.shapiro(data)
```

```python
count = np.array([12, 16, 18])
nobs = np.array([100, 120, 130])
chi2, p, _ = sm.proportion.proportions_chisquare(count, nobs)
print(chi2, p)
```

Do đó ta chấp nhận giả thuyết H và coi t 0 ỷ lệ phế
phẩm ở ba nhà máy là như nhau.

Với mức ý nghĩa 5% có thể cho rằng tình trạng hôn nhân của vợ và chồng là độc lập
với nhau hay không? 

$H_0$ : Tình trạng hôn nhân của chồng độc lập với vợ.

$H_1$ : Tình trạng hôn nhân của chồng không độc lập với vợ. 

```python
data = [[180, 34, 36], [58, 76, 54], [43, 34, 27]]
chi2, p, df, _ = ss.chi2_contingency(data)
print(chi2, p)
```

Do đó ta bác bỏ giả thuyết $H_0$
