1. Bài toán mô tả vật thể/điểm/vector trong không gian n-chiều.
Theo cách cũ, chúng ta biểu diễn một điểm A trong không gian 3 chiều (x, y, z) thông qua một tọa
độ có dạng, ví dụ: (12, 3, -5). Tương tự như vậy chúng ta có vector $\vec{OA}$ = (12, 3, −5) với O là
điểm gốc tọa độ. Mở rộng cách biểu diễn này trong không gian n chiều chúng ta có: 
 
  x = ($x_1$, $x_2$, ⋯ , $x_n$)
 
 Khi tiếp cận khái niệm ma trận, giả sử ta có ma trận $A_{m\times n}$ (m - hàng và n – cột), thông thường
chúng ta hình dung ma trận này được hình thành bằng cách xếp chồng m vector hàng lên với nhau
để hình thành ma trận $A_{m\times n}$. (Hoặc nếu chúng ta xây dựng ma trận này bằng cách xếp chồng các
vector cột với nhau thì sẽ có ma trận tương ứng là $A_{n\times m}$.) Phương pháp này có:
- **Ưu điểm:** là dùng một ma trận 2 chiều vẫn có thể quản lý được các vector trong không gian
có số chiều lớn hơn (> 3);
- **Nhược điểm:** việc quản lý đòi hỏi phải nắm bắt ý nghĩa hình thành ma trận (từ các vector
hàng hay cột); và bỏ qua cách quản lý từng chiều (từng trục tọa độ).
Vì vậy, chúng ta tiếp cận bài toán mô tả vật thể trong không gian n-chiều này theo 1 cách chính
quy hơn thông qua khái niệm Tensor.

**2. Tensor**


Tensor là một cấu trúc cho phép đặc tả số lượng chiều, giá trị của vật thể trong không gian n-chiều
này sẽ được truy cập thông qua việc truy cập tới các chỉ số của các chiều tương ứng.
1. Tensor hạng (rank) 0: các đơn trị, các đại lượng (vd: 15, -7, 9, .v.v.)
2. Tensor hạng 1: 1D array (vector x = (1, 2, 3, 4, 5) .v.v.) (truy cập tới phần tử thứ 2 của
vector trên trục tọa độ duy nhất của nó – print(x[1]) – 2)
3. Tensor hạng 2: 2D array – (tương đương ma trận 2 chiều).
4. Tensor hạng n: nth-D array – đặc tả đối tượng trong không gian n chiều.

**Bài tập 2 - Thư viện thao tác với cấu trúc Tensor**

* **TensorFlow**

```python
import tensorflow as tf # import tensor để tạo tensor 
```

* Tạo tensor hạng 1

```python
tensor_1 = tf.constant([1, 2, 3, 4 ,5]) # Tạo tensor bằng cách bình thường
tensor_1
```

* Tạo tensor hạng 2

```python
tensor_zeros_2=tf.zeros((3,3)) # Tạo tensor 0: tf.zeros; tensor đơn vị: tf.ones
tensor_zeros_2
```

* Tạo tensor hạng 3

```python
q=tf.range(start=0, limit=18, delta=1) # Tạo tensor hạng 1 với 18 phần tử
tensor_3 = tf.reshape(q,[2,3,3])  # Chuyển thành tensor hạng 3
tensor_3
```

* Tạo tensor hạng 4

```python
q=tf.range(start=0, limit=36, delta=1) # Tạo tensor hạng 1 với 36 phần tử
tensor_4 = tf.reshape(q,[2,2,3,3])  # Chuyển thành tensor hạng 4
tensor_4
```
